```python
from collections import defaultdict
import xml.etree.ElementTree as ET
```

```python
xml_file = '/Users/claire/Desktop/scraped_amp_replaced.xml'
tree = ET.parse(xml_file)

debate_id='ignore' # the first tags are not under a major-heading 
debates=defaultdict(list)

for elem in tree.iter(): # loop through all tags
    if elem.tag=='major-heading':
        print elem.attrib['id']
        debate_id='debate_%s' %elem.attrib['id'] # every time we hit a major-heading tag, we make a new key in the debates dictionary
    debates[debate_id].append(elem)
```

```python
for k,v in debates.iteritems():
    print k.upper()
    for i in v:
        print i.get('speakername', '').upper(), '\n',i.text
    
    print '---------------'
    
```
